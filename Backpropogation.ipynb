```
#Weights and Bias
w1=random.uniform(0, 1)
w2=random.uniform(0, 1)

w3=random.uniform(0, 1)

w4=random.uniform(0, 1)

w5=random.uniform(0, 1)

w6=random.uniform(0, 1)

w7=random.uniform(0, 1)

w8=random.uniform(0, 1)

b1=random.uniform(0, 1)

b2=random.uniform(0, 1)

```

```
n = 0.2

x = [[0,0,0],[1,0,1],[0,1,1],[1,1,1]]
y = x*50000
epoch=0
```

```
for i in y:
  out=0
  epoch=epoch+1
  if(epoch%400==0):
    print("Epoch:",(epoch/4))
  i1 = i[0]
  i2 = i[1]
  if i[2]==1:
    to1=1
    to2=0
    out=1
  else:
    to1=0
    to2=1
  #forward Pass
  neth1 = (w1*i1)+(w2*i2)+(b1*1)
  outh1 = 1/(1+math.exp(-neth1))

  neth2 = (w3*i1)+(w4*i2)+(b1*1)
  outh2 = 1/(1+math.exp(-neth2))

  neto1 = (w5*outh1)+(w6*outh2)+(b2*1)
  outo1 = 1/(1+math.exp(-neto1))

  neto2 = (w7*outh1)+(w8*outh2)+(b2*1)
  outo2 = 1/(1+math.exp(-neto2))
  Eo1 = 0.5*(to1-outo1)**2
  Eo2 = 0.5*(to2-outo2)**2
  Etotal = Eo1+Eo2

  #backward pass
  DEtotalOuto1 = -(to1-outo1)
  DOuto1Neto1  = outo1*(1-outo1)
  DNeto1W5     = outh1
  DNeto1W6     = outh2

  DEtotalW5 = DEtotalOuto1*DOuto1Neto1*DNeto1W5
  DEtotalW6 = DEtotalOuto1*DOuto1Neto1*DNeto1W6


  DEtotalOuto2 = -(to2-outo2)
  DOuto2Neto2  = outo2*(1-outo2)
  DNeto2W7     = outh1
  DNeto2W8     = outh2

  DEtotalW7 = DEtotalOuto2*DOuto2Neto2*DNeto2W7
  DEtotalW8 = DEtotalOuto2*DOuto2Neto2*DNeto2W8



  w5p = w5 - n*DEtotalW5
  w6p = w6 - n*DEtotalW6
  w7p = w7 - n*DEtotalW7
  w8p = w8 - n*DEtotalW8




  DEo1Outo1 = -(to1-outo1)
  DEo1Neto1 = DEo1Outo1*DOuto1Neto1
  DNeto1Outh1 = w5
  DEo1Outh1 = DEo1Neto1*DNeto1Outh1

  DEo2Outo2 = -(to2-outo2)
  DEo2Neto2 = DEo2Outo2*DOuto2Neto2
  DNeto2Outh1 = w7
  DEo2Outh1 = DEo2Neto2*DNeto2Outh1

  DEtotalOuth1 = DEo1Outh1+DEo2Outh1


  DOuth1Neth1 = outh1*(1-outh1)
  DNeth1W1 = i1
  DNeth1W2 = i2
  DEtotalW1 = DEtotalOuth1*DOuth1Neth1*DNeth1W1
  DEtotalW2 = DEtotalOuth1*DOuth1Neth1*DNeth1W2


  DNeto1Outh2 = w6
  DEo1Outh2 = DEo1Neto1*DNeto1Outh2
  DNeto2Outh2 = w8
  DEo2Outh2 = DEo2Neto2*DNeto2Outh2
  
  DEtotalOuth2 = DEo1Outh2+DEo2Outh2
  
  DOuth2Neth2 = outh2*(1-outh2)
  DNeth2W3 = i1
  DNeth2W4 = i2
  DEtotalW3 = DEtotalOuth2*DOuth2Neth2*DNeth2W3
  DEtotalW4 = DEtotalOuth2*DOuth2Neth2*DNeth2W4
  
  w1p = w1 - n*DEtotalW1
  w2p = w2 - n*DEtotalW2
  w3p = w3 - n*DEtotalW3
  w4p = w4 - n*DEtotalW4
  
  w1 = w1p
  w2 = w2p
  w3 = w3p
  w4 = w4p
  w5 = w5p
  w6 = w6p
  w7 = w7p
  w8 = w8p
```

```
 #forward Pass
x = [[0,0,0],[1,0,1],[0,1,1],[1,1,1]]

print("Predictions:")
print("X","Y","       F")
loss=0
for i in x:
  i1=i[0]
  i2=i[1] 
  neth1 = (w1*i1)+(w2*i2)+(b1*1)
  outh1 = 1/(1+math.exp(-neth1))

  neth2 = (w3*i1)+(w4*i2)+(b1*1)
  outh2 = 1/(1+math.exp(-neth2))

  neto1 = (w5*outh1)+(w6*outh2)+(b2*1)
  outo1 = 1/(1+math.exp(-neto1))
  neto2 = (w7*outh1)+(w8*outh2)+(b2*1)
  outo2 = 1/(1+math.exp(-neto2))
  loss = loss+abs(outo1-i[2])
  print(i1,i2,outo1,i[2]," Loss:",loss)
  
```
